### Implementing helper function for loading the files

```python
def get_web_data(url):
    response=requests.get(url)
    results=p.from_buffer(response.content)
    return results

def get_local_data(file_path):
    results=p.from_file(file_path)
    return results

def process_image_web(url):
    r = requests.get(url)
    img = Image.open(io.BytesIO(r.content))
    text = pytesseract.image_to_string(img)
    return text

def process_image_local(path):
    im = Image.open(path)
    text = pytesseract.image_to_string(im)
    return text
```
