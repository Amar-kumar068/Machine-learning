## Time Resampling

```python
## Rule-A --> Year-end frequency
data.resample(rule='A').max()
```

```python
data.resample(rule='A').max()['Open'].plot()
```

```python
## Quarterly- start frequency

data.resample(rule='QS').max()['High'].plot()
```

```python
## Business Year-end frequency

data.resample(rule='BA').max()['High'].plot()
```

```python
## Business Quarterly frequency 

data.resample(rule='BQS').max()['High'].plot()
```

```python
## Plotting

data['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
data['Open'].resample(rule='M').max().plot(kind='bar',figsize=(15,6))
```

## Simple moving Average

```python
data['Open'].plot(figsize=(15,6))
```

```python
data.head()
```

```python
data.drop('Open:no of days rolling', axis=1, inplace=True)
```

```python
data['Open:10 days rolling']=data['Open'].rolling(window=10, min_periods=1).mean()
```

```python
data[['Open','Open:10 days rolling']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,6))
```

```python
data['Open:30 days rolling']=data['Open'].rolling(window=30, min_periods=1).mean()
data['Open:50 days rolling']=data['Open'].rolling(window=50, min_periods=1).mean()
data[['Open','Open:10 days rolling','Open:30 days rolling','Open:50 days rolling']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,6))
```

## Expanding-- Cumulative Moving Average

```python
data['Open'].expanding().mean().plot(figsize=(10,5))
```

## Exponential moving average(EMA)

```python
## let smoothing factor=0.1
data['EMA_0.1']=data['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
data[['Open','EMA_0.1']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,6))
```

## Exponential weighted moving average(EWMA)


```python
data['EMA_5days']=data['Open'].ewm(span=5).mean()
```

```python
data[['Open','EMA_5days','EMA_0.1']].plot(figsize=(15,5))
```
